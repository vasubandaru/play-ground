```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0, 4*np.pi, 50)
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show()
```

```python
x = np.linspace(0, 4*np.pi, 50)
plt.plot(x, np.sin(x), color = "red", linewidth=3.0, linestyle='--', label = 'sin(x)')
plt.plot(x, np.cos(x), color = "black", label = 'cos(x)')

plt.xticks([0,np.pi,2*np.pi,3*np.pi,4*np.pi])
plt.yticks([-1,0,+1])

plt.legend(loc="upper right")
plt.show()
```

```python
x = np.arange(0, 10, 0.2)

# blue circles and green triangles
plt.plot(x, x**3, 'bo', x, x**2, 'g^')

plt.show()
```

```python
plt.subplot(224) # 2rows, 2cols, pos 4
plt.plot(x,np.sin(x), label='sin')
plt.legend()

plt.subplot(223) # 2rows, 2cols, pos 3
plt.plot(x,np.sin(x), label='sin')
plt.legend()

plt.subplot(222) # 2rows, 2cols, pos 2
plt.plot(x, np.cos(x), label = 'cos(x)')
plt.legend()

plt.subplot(221) # 2rows, 2cols, pos 1
plt.plot(x, np.cos(x), label = 'hahaha')
plt.legend()

plt.show()
plt.grid(True)
```
